plot channel profiles to show slope from 2016 DSM, along with channel profiles from bathymetric slope from 2000 and meander belt axis profiles from bathymetry

```python
# read csv files

import numpy as np
import pandas as pd

# dataframe for 2017 channel slopes
df1 = pd.read_csv('channel_slope_profs/channel_profs_2017.csv', header=0)

# dataframe for 2000 channel slopes
df2 = pd.read_csv('channel_slope_profs/channel_profs_2000_bath.csv', header=0)

# dataframe for meander belt axis slopes
df3 = pd.read_csv('channel_slope_profs/corrected_bath_meander_belt_axes.csv', header=0)
```

```python
# extract data for CM1 by referring to channel id

CM1_17 = df1['prof_id']== 1
CM1_CS17 = df1[CM1_17]

CM1_001 = df2['prof_id']== 1
CM1_CS00 = df2[CM1_001]

CM1_002 = df3['prof_id']== 1
CM1_mbaS00 = df3[CM1_002]

# channel and MBA slopes

Z_L_Ch = [0.017, 0.019, 0.017, 0.0053, 0.015, 0.0048, 0.012]
Z_L_MBA_unc = [0.03, 0.026, 0.024, 0.021, 0.02, 0.021, 0.027]
Z_L_MBA_c = [0.03, 0.027, 0.019, 0.01, 0.023, 0.014, 0.012]

# plot CM1 data on one axes plot

%matplotlib inline 
from matplotlib import pyplot as plt

ax1 = plt.axes()
#ax1.plot(CM1_CS00.cds2d, CM1_CS00.shore_bath, label='channel profile, 2000: Z/L = 0.028')
ax1.plot(CM1_mbaS00.cds2d, CM1_mbaS00.bathymetry, '--', label='meander belt axis profile, unc.: Z/L = 0.030')
ax1.plot(CM1_mbaS00.cds2d, CM1_mbaS00.bathymetry_new, '--', label='meander belt axis profile, c.: Z/L = 0.030')
ax1.plot(CM1_CS17.cds2d, CM1_CS17.DSM2016, color=[.6,.25,.31], label='channel profile, 2017: Z/L = 0.017')
plt.legend(loc='upper right')

plt.xlabel('distance [m]',fontsize=12)
plt.ylabel('elevation [m]',fontsize=12)
plt.title('Evolution of longitudinal topography, CM1', fontsize=14)

# label average slopes on figure
#plt.text(50,-422,"Z/L = 0.017",fontsize=10) #channel slope 2017
#plt.text(150,-412,"Z/L = 0.028",fontsize=10) #channel slope 2000
#plt.text(400,-418,"Z/L = 0.030",fontsize=10) #MBA slope 

# save figure
plt.savefig('CM1_channel-slope-evo.png', bbox_inches='tight')
plt.savefig('CM1_channel-slope-evo.svg')
```

```python
# plot data for CM1-3 and 6 in subplots by looping over each axis
%matplotlib inline 
from matplotlib import pyplot as plt

fig = plt.figure(figsize = (20,20))

# CM1-3
for n in range(1,4):
    CS17 = df1[df1.prof_id == n]
    #CS00 = df2[df2.prof_id == n]
    MBA = df3[df3.prof_id== n]
    ax = fig.add_subplot(3, 2, n)
    #ax.plot(CS00.cds2d, CS00.shore_bath, label='channel profile, 2000')
    ax.plot(MBA.cds2d, MBA.bathymetry, '--', label='meander belt axis profile, unc.: Z/L = %f' %Z_L_MBA_unc[(n-1)])
    ax.plot(MBA.cds2d, MBA.bathymetry_new, '--', label='meander belt axis profile, c.: Z/L = %f' %Z_L_MBA_c[(n-1)])
    ax.plot(CS17.cds2d, CS17.DSM2016, color=[.6,.25,.31], label='channel profile, 2017: Z/L = %f' %Z_L_Ch[(n-1)]) 
    plt.xlim(0,1300)
    plt.ylim(-431,-405)
    plt.xlabel('distance [m]',fontsize=12)
    plt.ylabel('elevation [m]',fontsize=12)
    plt.title('CM%d' %n)
    plt.legend(loc='upper right')

for n in range(6,7):
    CS17 = df1[df1.prof_id == n]
    #CS00 = df2[df2.prof_id == 4]
    MBA = df3[df3.prof_id == n]
    ax = fig.add_subplot(3, 2, n-2)
    #ax.plot(CS00.cds2d, CS00.shore_bath, label='channel profile, 2000')
    ax.plot(MBA.cds2d, MBA.bathymetry, '--', label='meander belt axis profile, unc: Z/L = %f' %(Z_L_MBA_unc[(n-2)]))
    ax.plot(MBA.cds2d, MBA.bathymetry_new, '--', label='meander belt axis profile, c: Z/L = %f' %(Z_L_MBA_c[(n-2)]))
    ax.plot(CS17.cds2d, CS17.DSM2016, color=[.6,.25,.31], label='channel profile, 2017: Z/L = %f' %(Z_L_Ch[(n-2)])) 
    plt.xlim(0,1300)
    plt.ylim(-431,-405)
    plt.xlabel('distance [m]',fontsize=12)
    plt.ylabel('elevation [m]',fontsize=12)
    plt.title('CM%d' %(n))
    plt.legend(loc='upper right')
    
for n in range(7,8):
    CS17 = df1[df1.prof_id == n]
    #CS00 = df2[df2.prof_id == 4]
    MBA = df3[df3.prof_id == n]
    ax = fig.add_subplot(3, 2, n-2)
    #ax.plot(CS00.cds2d, CS00.shore_bath, label='channel profile, 2000')
    ax.plot(MBA.cds2d, MBA.bathymetry, '--', label='meander belt axis profile, unc: Z/L = %f' %(Z_L_MBA_unc[(n-2)]))
    ax.plot(MBA.cds2d, MBA.bathymetry_new, '--', label='meander belt axis profile, c: Z/L = %f' %(Z_L_MBA_c[(n-2)]))
    ax.plot(CS17.cds2d, CS17.DSM2016, color=[.6,.25,.31], label='channel profile, 2017: Z/L = %f' %(Z_L_Ch[(n-2)])) 
    plt.xlim(0,650)
    plt.ylim(-431,-418)
    plt.xlabel('distance [m]',fontsize=12)
    plt.ylabel('elevation [m]',fontsize=12)
    plt.title('CM%d' %(n))
    plt.legend(loc='upper right')
    
for n in range(8,9):
    CS17 = df1[df1.prof_id == n]
    #CS00 = df2[df2.prof_id == 4]
    MBA = df3[df3.prof_id == n]
    ax = fig.add_subplot(3, 2, n-2)
    #ax.plot(CS00.cds2d, CS00.shore_bath, label='channel profile, 2000')
    ax.plot(MBA.cds2d, MBA.bathymetry, '--', label='meander belt axis profile, unc: Z/L = %f' %(Z_L_MBA_unc[(n-2)]))
    ax.plot(MBA.cds2d, MBA.bathymetry_new, '--', label='meander belt axis profile, c: Z/L = %f' %(Z_L_MBA_c[(n-2)]))
    ax.plot(CS17.cds2d, CS17.DSM2016, color=[.6,.25,.31], label='channel profile, 2017: Z/L = %f' %(Z_L_Ch[(n-2)])) 
    plt.xlim(0,300)
    plt.ylim(-431,-425)
    plt.xlabel('distance [m]',fontsize=12)
    plt.ylabel('elevation [m]',fontsize=12)
    plt.title('CM%d' %(n))
    plt.legend(loc='upper right')

plt.savefig('long_prof_channels.png', bbox_inches='tight')
plt.savefig('long_prof_channels.svg')    

plt.tight_layout()
plt.show()
```
